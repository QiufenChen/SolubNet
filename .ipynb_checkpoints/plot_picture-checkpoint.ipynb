```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

###  1. Cui9943

```python
CuiFile = "./dataset/Cui9943.csv"
csv_reader = csv.reader(open(CuiFile))
next(csv_reader)
data = []
for line in csv_reader:
    prop = float(line[3])
    data.append(prop)
CuiDF = np.array(data)
print("Max Value: ", np.max(CuiDF))
print("Min Value: ", np.min(CuiDF))
print("Mean Value: ", np.mean(CuiDF))
print("Median Value: ", np.median(CuiDF))
print("Numbers: ", len(CuiDF))

plt.hist(data,bins=100, label="Cui9943")
plt.xlabel("Exptl. logS", fontsize=10,fontweight='bold')
plt.ylabel("Frequency", fontsize=10,fontweight='bold')
plt.legend(loc='upper left',fontsize=10)
plt.savefig("./dataset/" + 'Cui.png', dpi=600)
plt.show()


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train_df = pd.read_csv("./dataset/random/CuiTrain.csv")
val_df = pd.read_csv("./dataset/random/CuiValid.csv")
test_df = pd.read_csv("./dataset/random/CuiTest.csv")
train = sorted(list(train_df.LogS))
valid = sorted(list(val_df.LogS))
test = sorted(list(test_df.LogS))

plt.figure(figsize=(7, 5), dpi=600)
plt.xlabel('LogS')
plt.ylabel('Density')
sns.kdeplot(train,shade=True,color='#FF2D2D', linewidth=1,
linestyle='--', label = 'CuiTrain')
sns.kdeplot(valid,shade=True,color='#02C874', linewidth=1,
linestyle='--', label = 'CuiValid')
sns.kdeplot(test,shade=True, color='#2894FF', linewidth=1,
linestyle='--', label = 'CuiTest')
plt.title("Random Split")
plt.legend()
plt.savefig("./dataset/random/Cui.png")
plt.show()
```

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
from scipy.stats import laplace
import warnings
from rdkit import Chem
from rdkit.Chem import Lipinski
from rdkit.Chem import Descriptors

data = pd.read_csv('DELAN.csv')
logp, molwt, logs = [], [], []
for i in range(len(data)):
    mol = Chem.MolFromSmiles(data.iloc[i,0])
    logp.append(Descriptors.MolLogP(mol))
    molwt.append(Descriptors.MolWt(mol))
    logs.append(data.iloc[i,1])
plt.figure(dpi = 600)
plt.xlabel('logP', fontproperties='KaiTi')
plt.ylabel('mol_weight', fontproperties='KaiTi')
plt.scatter(logp, molwt, s=3, c=logs, cmap='rainbow', vmin=-5, vmax=3, label='DELAN')
plt.colorbar(label="Logs", orientation="vertical",aspect=40)
plt.legend()
plt.show()
```
